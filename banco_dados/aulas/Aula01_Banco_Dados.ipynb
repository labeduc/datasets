## **Introdução a SQL**



*   A linguagem SQL é o recurso mais conhecido por DBAs e programadores para a execução de comandos em bancos de dados relacionais. É por meio dela que criamos tabelas, colunas, índices, atribuímos permissões a usuários, bem como realizamos consultas a dados. Enfim, é utilizando a SQL que “conversamos” com o banco de dados.

*   SQL = Structured Query Language = Linguagem Estruturada de Consulta


---

**Organização da Linguagem SQL:**
  *   Consulta dos dados: DQL - *Data Query Language*
  *   Manipulação dos dados: DML - *Data Manipulation Language*
  *   Manipulação da estrutura das tabelas: DDL - *Data Definition Language*
  *   Manipulação das permissões / control de acesso aos usuários: DDL - *Data Definition Language*
  *   Controle de acesso aos usuários: DDL - *Data Definition Language *   
  *   Controle de transação no banco de dados: DTL - *Data Transaction Language*

## **SELECT**

O comando SELECT permite consultar dados do banco de dados que estejam armazenados em tabelas, views (uma visualização de um subset de dados) ou stored procedures (funções ou procedimentos criados e compilados dentro do banco de dados).

**Sintaxe**: 
```
SELECT <lista_de_campos>
FROM <nome_da_tabela>

```

**Exemplo**:
```
SELECT codigo, nome
FROM clientes

```

## **Cláusula WHERE**


A cláusula Where permite ao comando SQL passar condições para filtrar dados.

Sintaxe:
```
SELECT <lista_de_campos>
FROM <nome_da_tabela>
WHERE <filtros>
```

Exemplo:
```
SELECT codigo, nome
FROM clientes
WHERE estado = 'RS'

```

Exemplo de outros operadores utilizados com a cláusula WHERE:


| **Operador**            | **Exemplo**
| ----------------------- | -------------------- 
| =                       | Select * from clientes where estado = ‘RS’
| <>                      | Select * from clientes where estado <> ‘RS’
| OR                      | Select * from clientes where estado = ‘RS’ OR estado = ‘SP’
| AND                     | Select * from clientes where estado = ‘RS’ AND cidade = ‘Porto Alegre’
| ()                      | Select * from clientes where (estado = ‘RS’ AND cidade = ‘Canoas’) OR (cidade = ‘Gravataí’)
| LIKE                    | Select * from clientes where nome LIKE ‘%Santos’
| NOT LIKE                | Select * from clientes where nome NOT LIKE ‘%Santos’
| IN                      | Select * from clientes where estado IN (‘RS’,’SP’)
| NOT IN                  | Select * from clientes where estado NOT IN (‘RS’,’SP’)
| BETWEEN                 | Select * from clientes where codigo BETWEEN 1 AND 3
| NOT BETWEEN             | Select * from clientes where codigo NOT BETWEEN 1 AND 3

## **Cláusula GROUP BY**

A cláusula group by permite agrupamento dos dados em um comando SQL.

**Sintaxe:**



```
SELECT <lista_de_campos>
FROM <nome_da_tabela>
WHERE <filtros>
GROUP BY <lista_de_campos>
```

**Exemplo:**



```
SELECT count(*),estado
FROM clientes
GROUP BY estado
```

*   Count - Contar número de linhas retornadas em uma consulta
*   Max - Traz o maior valor dentre as linhas retornas em uma consulta

## **Cláusula HAVING**

A cláusula HAVING determina uma condição de busca em conjunto com o GROUP BY.

**Sintaxe:**



```
SELECT <lista_de_campos>
FROM <nome_da_tabela>
WHERE <filtros>
GROUP BY <lista_de_campos>
HAVING <condicao>
```

**Exemplo:**



```
SELECT count(*),estado
FROM clientes
GROUP BY estado
HAVING count(*) > 1
```



````python
# Importar a biblioteca sqlite3 para podermos trabalhar com base de dados SQLite para praticar os exercícios da aula
#import sqlite3

# Download do dataset utilizado na aula - clientes.db
#!wget https://github.com/labeduc/datasets/raw/main/banco_dados/clientes.db

# Mostrando o conteúdo da tabela clientes
#cursor = conn.execute("SELECT * FROM Clientes")
#for row in cursor:
#  print(row)

# Ou utilizando esta outra forma
#%load_ext sql
#%sql sqlite:///clientes.db
````
